#  Univariate Analysis

# **Ctegorigal column analysis**

```python

# plot of the output column
plt.pie(df['loan_status'].value_counts(), labels=['0', '1'], autopct='%1.1f%%', colors=['skyblue', 'orange'])
plt.title('Distribution of Output Column')
plt.show()
# This is imbalanced dataset problem
# number of positivre class is less than number of negetive class


# Pie char plot on gender column
plt.pie(df['person_gender'].value_counts(), labels=['Male', 'Female'], autopct='%1.1f%%', colors=['lightgreen', 'lightpink'])
plt.title(' Gender Pie chat')
plt.show()

# Pie chat on person eduction
plt.pie(df['person_education'].value_counts(), labels=['Bachelor', 'Associate','High School','Master','Doctorate'], autopct='%1.1f%%', colors=['lightgreen', 'lightpink','skyblue','yellow','red'])
plt.title(' Person Eduction Pie chat')
plt.show()

# Pie chat on loan intent

plt.pie(df['loan_intent'].value_counts(), labels=['Education', 'Medical','Venture','Personal','Debt Consolidation','Home Improvement'], autopct='%1.1f%%', colors=['lightgreen', 'lightpink','skyblue','orange','yellow','red'])
plt.title(' Person intent for loan  Pie chat')
plt.show()
```

# **Numerical column aanalysis**

```python
# Plot histogram on the age
plt.hist(df['person_age'],bins=20,color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram for age ')
plt.show()
print('Max Age : ',df['person_age'].max())


# plot Box plot on age column
plt.boxplot(df['person_age'])
plt.ylabel('Age')
plt.title('Box plot for age ')
plt.show()
```
