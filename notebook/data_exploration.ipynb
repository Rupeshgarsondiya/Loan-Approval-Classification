**1 . Age Analysis**

```python
# Plot histogram on the age
plt.hist(df['person_age'],bins=20,color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram for age ')
plt.xlim(0,150)
plt.show()
print('Max Age : ',df['person_age'].max())

# plot distplot
sns.displot(df['person_age'],kde=True,color='skyblue')
plt.xlim(0,100)
plt.show()

# plot Box plot on age column
'''
By Visulize the boxplot in our age column outlier are present in our dataset
'''
plt.boxplot(df['person_age'])
plt.ylabel('Age')
plt.title('Box plot for age ')
plt.show()

sns.kdeplot(df['person_age'],color='red',shade=True)
plt.show()

print('Skew factor of age column : ',df['person_age'].skew())

'''
Conclusion :  This checks whether the graph is symmetric or not. If the value is zero, it means the graph is normally distributed. Therefore, our age column is not
              normally distributed.
'''
```
